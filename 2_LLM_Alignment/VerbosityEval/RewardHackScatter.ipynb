```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data
df_dpo = pd.read_csv("eval_results_DPO.csv")
df_ppo_s = pd.read_csv("eval_results_PPO-Sparse.csv")
df_sft = pd.read_csv("eval_results_SFT-Baseline.csv")
# df_grpo = pd.read_csv("eval_results_GRPO.csv") # Add later

# Combine
df_all = pd.concat([df_dpo, df_ppo_s, df_sft])

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_all, x="Length (Tokens)", y="Reward_Score", hue="Model", alpha=0.8)
plt.title("Reward Hacking Check: Length vs Score")
plt.axhline(y=0, color='gray', linestyle='--')
plt.show()
```
